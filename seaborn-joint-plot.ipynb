```python
# Setup
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Load data
# Tips
tips = sns.load_dataset('tips')

tips.head()
```

```python
# Joint plot tips
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'reg')
```

```python
# Change the kind attribute
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'kde')
```

```python
# Change stat_func attribute
from scipy.stats import spearmanr

sns.jointplot(x = 'size', y = 'total_bill', data = tips, stat_func = spearmanr)
```

```python
# Change size attribute
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, ratio = 4, height =6)
```

```python
# Load data
# Iris
iris = sns.load_dataset('iris')

iris.head()
```

```python
# Joint plot iris
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'resid')
```

```python
# Change the kind attribute
sns.jointplot(x = 'sepal_length', y = 'sepal_width', data = iris, kind = 'kde')
```
